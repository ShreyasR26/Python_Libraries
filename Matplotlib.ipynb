```python
X=np.linspace(0,20,100)
A_signal=np.sin(X)
B_signal=np.cos(X)
#print(X)
plt.plot(X,A_signal,label='Sine Wave',color='red')
plt.plot(X,B_signal,label='Cos Wave', color='blue')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

```python
#XOR scatter plot
L=5000
X=np.random.rand(L,2)*2 - 1
Y=np.zeros(L)
for i in range(0,L):
    if(X[i][0]>0 and X[i][1]>0 or X[i][0]<0 and X[i][1]<0):
        Y[i]=1
    else:
        Y[i]=0
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```
