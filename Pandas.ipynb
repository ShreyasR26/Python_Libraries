```python
def getflower(sl,sw,pl,pw):
    sa=sl*sw
    pa=pl*pw
    diff1=abs(11.29-pa)
    diff2=abs(0.36-pa)
    diff3=abs(5.72-pa)
    if(diff3<=diff2 and diff3<=diff1):
        return 'Iris-versicolor'
    if(diff2<=diff1 and diff2<=diff3):
        return 'Iris-setosa'
    if(diff1<=diff2 and diff1<=diff3):
        return 'Iris-virginica'
```

```python
df['flower_predicted']=[getflower(a,b,c,d)for a,b,c,d in zip(df.sepal_length,df.sepal_width, df.petal_length,df.petal_length)]
```
